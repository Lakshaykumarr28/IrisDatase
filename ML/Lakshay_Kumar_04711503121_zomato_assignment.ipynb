Ques. Plot piechart for Country vs Average cost for two

```python
labels1 = df3.Country.value_counts().index
values = df3.Country.value_counts().values

plt.pie(values , labels = labels1, autopct = "%1.1f%%")
```

Ques. Plot piechart for highest three countries.

```python
plt.pie(values[:3], labels = labels1[:3], autopct = "%1.1f%%")
```

Ques. Groupby aggregate rating, rating color, rating text.

```python
df4 = df3.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating Count'})
df4
```

Ques. Bar plot aggregate rating vs rating count.

```python
sns.barplot(data = df4 , x = "Aggregate rating", y = "Rating Count")
```

Ques. Count plot rating color vs rating count.

```python
sns.countplot(data = df4, x = 'Rating color', hue='Rating Count')
```

Ques. Which currency is used by which country.

```python
df5 = df3.groupby(['Country', 'Currency']).size().reset_index()
df5
```

Ques. Which country have online delivery options?

```python
df6 = df3.groupby(['Country', 'Has Online delivery']).size().reset_index()
df6 = df6[df6['Has Online delivery'] == 'Yes']
df6
```
