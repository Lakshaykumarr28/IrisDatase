q1: From the given dataset print the first and last five rows

```python
#Solution
df2  = df.iloc[:5]
print(f"the first five rows are: \n{df2}")

df3  = df.iloc[-5:]
print(f"the last five rows are: \n{df3}")
```

q2: Clean the dataset and update the CSV file. Replace all column values which contain ?, n.a, or NaN with ?

```python
#Solution
df4 = df.fillna(0)
print(f"after cleaning the database, removing all NaN values: \n{df4}")
```

q3: Print most expensive car’s company name and price.

```python
#Solution
max = df['price'].max()
df5 = df[df['price'] == max]
df5
```

q4: Print All Toyota Cars details

```python
df5 = df[df['company'] == 'toyota']
print(f"all toyota car details: \n{df5}")
```

q5: Count total cars per company

```python
#Solution
df6 = df['company'].value_counts()
print(f"the number of models per company: \n{df6}")
```

q6: Find each company’s Highest price car

```python
#Solution
df2 = df.groupby('company')
df4 = df2['company', 'price'].max()
df4
```

q7: Find the average mileage of each car making company

```python
#Solution
df2 = df.groupby('company')
df4 = df2['company', 'average-mileage'].mean()
df4
```

q8: Sort all cars by Price column

```python
#Solution
df2 = df.sort_values(['price'])
df2
```
