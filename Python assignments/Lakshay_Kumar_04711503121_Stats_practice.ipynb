Ques. Read csv files and find mean, median, mode.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Summer Training/marks student.csv')
```

```python
df2 = df['mark'].mean()
print(f'mean of the data is: {df2}')
df3 = df['mark'].median()
print(f'median of the data is: {df3}')
df4 = df['mark'].mode()
print(f'mode of the data is: {df4}')
```
