```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Summer Training/company_sales_data.csv')
```

q1: Read Total profit of all months and show it using a line plot

    X label name = Month Number
    
    Y label name = Total profit

```python
#Solution
monthNames = df['month_number']
totalProfit = df['total_profit']

plt.plot(monthNames, totalProfit)
plt.xlabel('Month number')
plt.ylabel('Total Profit')
plt.show()
```

q2: Get total profit of all months and show line plot with the following Style properties

    Line Style dotted and Line-color should be red

    Show legend at the lower right location.

    X label name = Month Number

    Y label name = Sold units number

    Add a circle marker.

    Line marker color as read
    
    Line width should be 3

```python
#Solution
monthNames = df['month_number']
totalProfit = df['total_profit']

plt.plot(monthNames, totalProfit, 'r:', label = 'Sold Units number', linewidth = 3, marker = '*')
plt.xlabel('Month number')
plt.ylabel('Sold Units Number')
plt.legend(loc = 'lower right')
plt.show()
```

q3: Read all product sales data and show it  using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

```python
#Solution
df2 = df['facecream']
df3 = df['facewash']
df4 = df['toothpaste']
df5 = df['bathingsoap']
df6 = df['shampoo']
df7 = df['moisturizer']

x = np.arange(1,13,1)
plt.xlabel('Month Number')
plt.ylabel('Sold Number units')
plt.plot(x,df2,'r',  x,df3,'b',  x,df4,'g',   x,df5,'y',   x,df6,'orange',   x,df7,'black')
```

q4: Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.

```python
#Solution
plt.scatter(x,df4)
plt.grid(linestyle = '--')
```

q5: Read face cream and facewash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
#Solution
plt.bar(x-0.1,df2, color = 'blue', width = 0.2)
plt.bar(x+0.1,df3, color = 'red', width = 0.2)

plt.xlabel('product')
plt.ylabel('units sold per month')
```

q6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

```python
#Solution
plt.bar(x,df5, label = 'bathing soap')

plt.xlabel('month number')
plt.ylabel('sales per month')
plt.legend()
plt.show()

plt.savefig('myplot.png')
```

q7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

q8: Calculate total sale data for last year for each product and show it using a Pie chart

```python
#Solution
df2 = df['facecream'].sum()
df3 = df['facewash'].sum()
df4 = df['toothpaste'].sum()
df5 = df['bathingsoap'].sum()
df6 = df['shampoo'].sum()
df7 = df['moisturizer'].sum()
values = [df2,df3,df4,df5,df6,df7]
pro_labels = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']

plt.pie(values, labels = pro_labels, autopct = '%1.1f%%')
```

q9: Read Bathing soap facewash of all months and display it using the Subplot

```python
#Solution
x = np.arange(1,13,1)
df5 = df['moisturizer']

plt.subplot(2,3,1)
plt.xlabel('month number')
plt.ylabel('units sold')
plt.plot(x,df5)
```

q10: Read all product sales data and show it using the stack plot

```python
#Solution
df2 = df['facecream']
df3 = df['facewash']
df4 = df['toothpaste']
df5 = df['bathingsoap']
df6 = df['shampoo']
df7 = df['moisturizer']
y = np.stack([df2,df3,df4,df5,df6,df7])
pro_labels = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']

fig, ax = plt.subplots()
ax.stackplot(x,y)
```
