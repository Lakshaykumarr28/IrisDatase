Q3: Check Palindrome
Given a string, determine if it is a palindrome, considering only alphanumeric characters.

```python
#Write Solution Here
str1 = input("enter string: ")
str2 = str1[::-1]

if(str1 == str2):
  print(f"the string {str1} is palindrome")
else:
  print(f"the string {str1} is not palindrome")
```

Q4:
Consider the string str=”Global Warming” Write statements in Python to implement the following

(a) To display the last four characters.

(b) To display the substring starting from index 4 and ending at index 8.

(c) To check whether string has alphanumeric characters or not

(d) To trim the last four characters from the string.

(e) To trim the first four characters from the string.

(f) To display the starting index for the substring „ WaD.

(g) To change the case of the given string.

(h) To check if the string is in title case.

(i) To replace all the occurrences of letter „aD in the string with „*?

```python
#Write Solution Here
str = "Global Warming"
print(f"(a) the last 4 characters are: {str[-4::]}")

print(f"(b) the substring starting from index 4 and ending at index 8: {str[4:8:]}")

print(f"(c) is the string alphanumeric: {str.isalnum()}")

str2 = str[0:-4]
print(f"(d) trimming the last 4 characters from the string: {str2}")

str3 = str[4:]
print(f"(e) trimming the first 4 characters from the string: {str3}")

print(f"(f) index of substring starting from 'W' is: {str.find('W')}")

str4 = str.lower()
str5 = str.upper()
print(f"(g) string in lower case: '{str4}' and string in upper case: '{str5}'")

print(f"(h) is the string in title case: {str.isupper()}")

str6 = str.replace("a", "*")
print(f"(i) replacing all 'a' with '*' gives: {str6}")
```

Ques. Take a list of integers and find the count of odd integers and even integers

```python
num = [0,0,0,0,0]
odd_count = 0
even_count = 0
for x in range(5):
  num[x] = int(input(f"enter number {x}: "))

for x in range(5):
  if(num[x] %2 != 0):
    odd_count +=1
  else:
    even_count += 1

print(f"even numbers: {even_count} \nodd numbers: {odd_count}")
```
